```python
import torch
```

```python
X = torch.tensor([1, 2, 3, 4], dtype=torch.float32)
Y = torch.tensor([5,10,15,20], dtype=torch.float32)
w1 = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
w0 = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
```

```python
def forward(x):
    return w1 * x + w0

def loss(y, y_pred):
    return torch.pow((y- y_pred), 2).mean()

```

```python
epoch = 5000
lr = 0.01

for i in range(epoch):
    y_pred = forward(X)

    cost = loss(Y, y_pred)

    cost.backward()

    with torch.no_grad():
        w1 -= lr * w1.grad
        w0 -= lr * w0.grad

    w1.grad.zero_()
    w0.grad.zero_()

    print(f'epoch: {i + 1}, output: {y_pred}, loss: {cost}')
```
